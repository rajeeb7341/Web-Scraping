```
import requests
```

```
url = 'https://www.python.org/jobs/'
req = requests.get(url)
```

```
req
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(req.text, 'lxml')
jobs = soup.find('ol',{'class':'list-recent-jobs'})
```

```
jobs.findAll('li')
```

```
for job in jobs.findAll('li'):
  d = dict()
  d['job_title'] = job.find('h2').find('a').text
  d['job_location'] = job.find('span',{'class':'listing-location'}).text
  d['job_date'] = job.find('span',{'class':'listing-posted'}).find('time').text

  print(d)
```

```
import pandas as pd
df = pd.DataFrame()
df['job_title'] = [job.find('h2').find('a').text for job in jobs.findAll('li') ]
df['job_location'] = [job.find('span',{'class':'listing-location'}).text for job in jobs.findAll('li')]
df['job_date'] = [job.find('span',{'class':'listing-posted'}).find('time').text for job in jobs.findAll('li')] 
df
```
